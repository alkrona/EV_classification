```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/EV_data.csv")
df.head(20)
```

```python
df['read_date'] = df['read_date'].str.replace(' 0:00', '')
df['read_date'] = pd.to_datetime(df['read_date'], format='%m/%d/%Y')

print(df["read_date"].dtype)
```

```python
df.to_csv("../data/EV_data_clean.csv")
```

```python
df_has_ev = df[(df['read_date']=='2021-03-01') & (df['label']==1)]
df_has_ev_top10=df_has_ev[:10]
df_has_ev_top10.shape
```

```python
df_has_no_ev = df[(df['read_date']=='2021-03-01') & (df['label']==0)]
df_has_no_ev_top10=df_has_no_ev[:10]
df_has_no_ev_top10.shape
```

```python
df_has_ev_top10.head()
```

```python
trial_y_values = df_has_no_ev_top10.iloc[:,1:49]
trial_y_values.shape
trial_x_values = df_has_ev_top10.columns[1:49]
```

```python
sns.lineplot(x=trial_x_values.values, y=trial_y_values.values)  # Use index for x-axis

# Optional customization
plt.xlabel("Column Index (2nd to 49th)")
plt.ylabel("Values")
plt.title("Line Chart from First Row (2nd to 49th Columns)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have 10 sets of data (data1, data2, ..., data10)

# Create a figure with enough height for 10 plots
plt.figure(figsize=(8, 15))  # Adjust figsize for readability

# Loop through each data set and create a subplot
for i in  range(trial_y_values.shape[0]):
    
    # Extract x and y data from your data structure (assuming data is a DataFrame with x and y columns)
    y_data = trial_y_values.iloc[i,:]  # Replace with your actual x-column name
    x_data = trial_x_values.values  # Replace with your actual y-column name
    
    # Create a subplot on each iteration
    ax = plt.subplot(10, 1, i+1)  # Create subplots in 10 rows, 1 column, starting from row 1
    
    # Create the line plot with Seaborn
    sns.lineplot(x=x_data, y=y_data, ax=ax)  # Specify the subplot (ax)

    # Optional: Customize each plot (titles, labels, etc.)
    ax.set_title(f"Plot {i+1}")

# Adjust layout (optional)
plt.tight_layout()

plt.show()
```
