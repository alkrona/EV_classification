```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/EV_data_clean.csv')
df.head()
```

```python
df = df.drop(['label','read_date','Unnamed: 0'],axis=1)
```

```python
df.head()
```

```python
grouped = df.groupby(['id']).mean()
```

```python
grouped.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, adjusted_rand_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
n_clusters = 4
preprocessor = Pipeline(
    [
        ("scaler", MinMaxScaler()),
        ("pca", PCA(n_components=4, random_state=42)),
    ]
)
```

```python
clusterer = Pipeline(
   [
       (
           "kmeans",
           KMeans(
               n_clusters=n_clusters,
               init="k-means++",
               n_init=50,
               max_iter=500,
               random_state=42,
           ),
       ),
   ]
)
```

```python
pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)
```

```python
data = grouped
pipe.fit(data)
```

```python
preprocessed_data = pipe["preprocessor"].transform(data)

predicted_labels = pipe["clusterer"]["kmeans"].labels_

silhouette_score(preprocessed_data, predicted_labels)
```

```python
predicted_labels[:10]
```

```python
grouped['power-usage-class']= predicted_labels
```

```python
grouped.to_csv('../data/EV_data_clusters_added')
```

```python
sse=[]
silhouette_coefficients = []
for k in range(2,11):
    n_clusters = k
    preprocessor = Pipeline(
        [
            ("scaler", MinMaxScaler()),
            ("pca", PCA(n_components=4, random_state=42)),
        ]
    )
    clusterer = Pipeline(
    [
        (
            "kmeans",
            KMeans(
                n_clusters=n_clusters,
                init="k-means++",
                n_init=50,
                max_iter=500,
                random_state=42,
            ),
        ),
    ]
    )
    pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
    )
    data = grouped
    pipe.fit(data)
    sse.append(pipe["clusterer"]["kmeans"].inertia_)
    score = silhouette_score(data,pipe["clusterer"]["kmeans"].labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), sse)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
